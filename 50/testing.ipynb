```python
# Investing distribution of numeric data

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_excel('practice_data.xlsx')

numeric_columns = ['Days Present', 'Sales Revenue Generated']
fig, axes = plt.subplots(1, 2, figsize=(12, 6))
for index, column in enumerate(numeric_columns):
    sns.kdeplot(data[column], ax=axes[index])
    axes[index].set_title(f'The title for the KDE plots by myself')
plt.tight_layout()
plt.show()
```

```python
# Shapiro Wilk test of normality

from scipy.stats import shapiro
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_excel('practice_data.xlsx')


numeric_columns = ['Days Present', 'Sales Revenue Generated']
shapiro_results = {}

for column in numeric_columns:
    stat, p_value = shapiro(data[column])
    shapiro_results[column] = round(p_value, 3)

shapiro_results


```

```python
# Starting with Square root transformation 

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

data = pd.read_excel('practice_data.xlsx')

def sqrt_transformation(data, column_name):
    data[f'{column_name}_sqrt'] = np.sqrt(data['Sales Revenue Generated'])
    stat, p_value = shapiro(data[f'{column_name}_sqrt'])
    # kdeplot = sns.kdeplot(data[f'{column_name}_sqrt'])
    distribution = sns.kdeplot(data[f'{column_name}_sqrt'])
    print(distribution)
    print('P value: ', p_value)

sqrt_transformation(data , "Sales Revenue Generated")
```

```python
# Logarithmic transformation method 

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_excel('practice_data.xlsx')


def log_transformation(data, column_name):
    data[f'{column_name}_log'] = np.log(data[column_name])
    stat, p_value = shapiro(data[f'{column_name}_log'])
    distribution = sns.kdeplot(data[f'{column_name}_log'])
    print(distribution)
    print('p-value: ', p_value)

log_transformation(data , 'Sales Revenue Generated')
```
